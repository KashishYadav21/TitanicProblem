Handling missing data

```python
 df=df.drop(columns='Cabin',axis=1)
```

Finding the mean of age coloumn and replace it by null value

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

embarked is categorical data type sowe find mode for this

```python
print(df['Embarked'].mode())
```

```python
print(df['Embarked'].mode()[0])
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

Encoding the Categorical column

```python
df['Sex'].value_counts()
```

```python
df['Embarked'].value_counts()
```

```python
df.replace({'Sex':{'male':0,'female':1 },'Embarked':{'S':0,'C':1,'Q':2}},inplace=True)
```

```python
df.head()
```

Separating feature and target column

```python
x=df.drop(columns=['PassengerId','Name','Ticket','Survived'],axis=1)
y=df['Survived']
```

```python
print(x)
```

```python
print(y)
```

splitiing the data into train and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

```python
lgr=LogisticRegression()
```

```python
lgr.fit(x_train,y_train)
```

```python
y_train_predict=lgr.predict(x_train)
```

calculate the accuracy score of predict and actual one

```python
acuuracy_train=accuracy_score(y_train_predict,y_train)
```

```python
print('Accuracy of trainig data is',acuuracy_train)
```

```python
y_test_predict=lgr.predict(x_test)
```

```python
acuraccy_test=accuracy_score(y_test,y_test_predict)
```

```python
print('Accuracy of test data is:', acuraccy_test)
```
